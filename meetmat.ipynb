```python
import whisper
from transformers import pipeline, AutoTokenizer, AutoModelForQuestionAnswering
from deep_translator import GoogleTranslator
from gtts import gTTS
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import gradio as gr
import torch
import os
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

# Load models
asr_model = whisper.load_model("base")
summarizer = pipeline("summarization", model="t5-small")
tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")
qa_model = AutoModelForQuestionAnswering.from_pretrained("bert-large-uncased-whole-word-masking-finetuned-squad")

# Email function
def send_email_summary(subject, body, recipients):
    sender_email = "abhishekk6593@gmail.com"
    sender_password = "wjvz qvva wqyp bkvp"

    for recipient in recipients:
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = recipient
        msg['Subject'] = subject

        msg.attach(MIMEText(body, 'plain'))

        with smtplib.SMTP('smtp.gmail.com', 587) as server:
            server.starttls()
            server.login(sender_email, sender_password)
            server.send_message(msg)

# MeetMate logic
def meetmate_pipeline(audio, question=None, emails="", target_lang="hi"):
    result = asr_model.transcribe(audio)
    transcript = result['text']

    # Summarize
    summary = summarizer(transcript, max_length=150, min_length=30)[0]['summary_text']

    # Translate summary
    try:
        translated = GoogleTranslator(source='auto', target=target_lang).translate(summary)
    except:
        translated = "Translation failed. Language might not be supported."

    # Text-to-Speech output of summary
    tts = gTTS(text=summary, lang='en')
    tts.save("summary.mp3")

    # Q&A Module
    answer = ""
    if question:
        inputs = tokenizer(question, transcript, return_tensors='pt')
        with torch.no_grad():
            outputs = qa_model(**inputs)
            answer_start = torch.argmax(outputs.start_logits)
            answer_end = torch.argmax(outputs.end_logits) + 1
            answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs['input_ids'][0][answer_start:answer_end]))

    # Clustering using TF-IDF
    sentences = transcript.split('.')
    tfidf = TfidfVectorizer(stop_words='english')
    X = tfidf.fit_transform(sentences)
    kmeans = KMeans(n_clusters=2, random_state=42).fit(X)
    topics = {i: [] for i in range(2)}
    for idx, label in enumerate(kmeans.labels_):
        topics[label].append(sentences[idx].strip())

    clustered = "\n\n".join([f"Topic {i+1}:\n" + "\n".join(sents) for i, sents in topics.items()])

    # Email summary
    if emails:
        recipients = [email.strip() for email in emails.split(',') if email.strip()]
        body = f"Meeting Summary:\n\n{summary}\n\nKey Topics:\n{clustered}"
        send_email_summary("MeetMate: Meeting Summary", body, recipients)

    return transcript, summary, translated, answer, clustered, "summary.mp3"

# Gradio Interface with Language Dropdown
iface = gr.Interface(
    fn=meetmate_pipeline,
    inputs=[
        gr.Audio(type="filepath", label="🎧 Upload Meeting Audio"),
        gr.Textbox(label="❓ Ask a Question (Optional)"),
        gr.Textbox(label="📧 Participant Emails (comma-separated)"),
        gr.Dropdown(
            label="🌐 Select Translation Language",
            choices=["hindi", "french", "spanish", "telugu", "tamil", "german", "bengali", "maliyali", "gujarati", "japanese", "korean", "chiniese"],
            value="hi"  # Default is Hindi
        )
    ],
    outputs=[
        gr.Textbox(label="📄 Transcript", lines=6),
        gr.Textbox(label="📝 Summary", lines=3),
        gr.Textbox(label="🌍 Translated Summary", lines=3),
        gr.Textbox(label="🎯 Q&A Result", lines=2),
        gr.Textbox(label="🧠 Clustered Topics", lines=6),
        gr.Audio(label="🔊 Summary Audio")
    ],
    title="🤖 MeetMate: AI Meeting Assistant",
    description="Upload a meeting audio file to get a full transcript, summary, translated summary, Q&A, topic clusters, audio summary, and auto-email summaries."
)

iface.launch(share=True)
```
